```python
%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('data/C2A2_data/BinnedCsvs_d400/fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89.csv')
df = df[~df['Date'].isin(['2008-02-29','2012-12-29'])] #Remoção das datas de ano bissexto
df['Day'] = pd.DatetimeIndex(df['Date']).day
df2 = df[df['Date'] <= '2014-12-31'] #Anteriores a 2015
df3 = df[df['Date'] > '2014-12-31'] #Posteriores a 2015

df_min = df2[df2['Element'] =='TMIN'].groupby('Day').aggregate({'Data_Value':np.min})
df_max = df2[df2['Element'] == 'TMAX'].groupby('Day').aggregate({'Data_Value':np.max})

plt.plot(df_max,'-o',df_min,'-o')

values = df_max.index.values
plt.fill_between(values,df_min.values.flatten(),df_max.values.flatten(), color='grey',alpha='0.5')

df3_max = df3[df3['Data_Value'] > df3['Day'].map(df_max['Data_Value'])]
df4_min = df3[df3['Data_Value'] < df3['Day'].map(df_min['Data_Value'])]
plt.scatter(df3_max['Day'].tolist(),df3_max['Data_Value'],s=10,c='red') #values over max
plt.scatter(df4_min['Day'].tolist(),df4_min['Data_Value'],s=10,c='black') #values under min

plt.xticks(list(df_max.index),[x for x in range(0,31)]) #we want to show all the x values and rename according to the month
plt.title('Min and Max temperatures in United States between 2005-2015', fontsize=20)
plt.xlabel('Years', fontsize=20)
plt.ylabel('Temperatures', fontsize=20)
fig = plt.gcf()
fig.set_size_inches(20,10)
plt.tick_params(labelsize=20)
```
